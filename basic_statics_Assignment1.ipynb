```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data=pd.read_csv('/content/sales_data_with_discounts.csv')
data
```

```python
numerical_columns = data.select_dtypes(include=['number']).columns
numerical_columns
```

**Interpretation**

Mean: Provides an overall average of sales and discounts.

Median: Indicates the central tendency, which is less affected by outliers.

Mode: Shows the most common values in the dataset.

Standard Deviation: Indicates how spread
out the values are, helping to identify variability in sales and discounts.

```python
statistics = data[numerical_columns].agg(['mean', 'median', lambda x: x.mode()[0], 'std'])
statistics.index = ['Mean', 'Median', 'Mode', 'Standard Deviation']
print(statistics)
```

**Data** **Visualization**


The graphical representation of data and information




**Histograms**

```python
import matplotlib.pyplot as plt
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=10)
    plt.title(f'Histogram of {column}')
    print('The min value of Avg Price in sales data is :',data[column].min())
    print('The max value of Avg Price in sales data is :',data[column].max())
    print('The skewness of Discount Amount in sales data is:',data[column].skew())
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```

**Boxplots**

```python
for column in numerical_columns:
    plt.figure(figsize=(8,6))
    plt.boxplot(data[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.grid(False)
    plt.show()
```

**Standardization** **Numerical** **variables**

```python
from sklearn.preprocessing import StandardScaler

#Standardize numerical columns
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data[numerical_columns])
standardized_df = pd.DataFrame(standardized_data, columns=numerical_columns)

#Show before and after comparisons
print("Before Standardization:")
print(data[numerical_columns].describe())
print('\n')
print("\nAfter Standardization:")
print(standardized_df.describe())
```
