```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_excel("/content/Online retail.xlsx")
print(df.head())
```

```python
#data preprocessing
transacttions=df.iloc[:,0].apply(lambda x: x.split(','))
for i in range(len(transacttions)):
  transacttions[i]=[j.strip() for j in transacttions[i]]
```

```python
#one-hot encode the data
te = TransactionEncoder()
te_ary = te.fit(transacttions).transform(transacttions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
df_encoded.head()
```

```python
df.replace({True:1, False:0}, inplace=True)
df.head()
```

```python
# Apriori Algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
print("frequent_itemsets:")
print(frequent_itemsets.sort_values(by='support', ascending=False).head(10))
```

```python
#Association Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
filtered_rules = rules[(rules['confidence'] > 0.5) & (rules['lift'] > 1.2)]
print("Top Association Rules:")
print(filtered_rules.sort_values(by='lift', ascending=False).head(10))
```

```python
for index, row in filtered_rules.sort_values(by="lift",ascending=False).head(5).iterrows():
  print(f"Rule {index+1}:")
  print(f"antecedents: {', '.join(row['antecedents'])}")
  print(f"consequents: {', '.join(row['consequents'])}")
  print(f"support: {row['support']}")
  print(f"confidence: {row['confidence']}")
  print(f"lift: {row['lift']}")
  print()
```

**1. What is Lift and Why is it Important in Association Rules?**

-->Lift measures how much more likely two items are bought comparesd to beign bought independently.

Lift is important because it helps identify strong associations and it shows how much more likely two items are purchased together than expected if they were independent.


**2. What is support and Confidence. How do you calculate them?**

-->Support measures the relative frequency of an itemset in a dataset,indicating how often a particular itemset appears.

Support(X) = Number of transactions containing itemset X /Total number of transctions

-->Confidence indicates the likelihood that item Y appears in a transaction given that item X is present.

Confidence(X -> Y) = support(XandY) / support(X)
